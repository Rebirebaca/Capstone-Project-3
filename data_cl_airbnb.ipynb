```python
#hotel infprmation:
def hotel_info(airb_data):
        h_data = {'id':[],'Listing_url':[],'Name':[], 'Description':[],'Neighborhood_Overview':[],'Transit':[],'Street':[],'Suburb':[],
        'Government_area':[],'Longitude':[],'Latitude':[],'Market':[],'Country':[],'Country_code':[],'Price':[],'Security_deposit':[],
        'Weekly_price':[],'Monthly_price':[],'Cleaning_fee':[],'Cancellation_policy':[]}

        for i in airb_data:
                ids = i['_id']
                h_data['id'].append(ids)
                listingu = i['listing_url']
                h_data['Listing_url'].append(listingu)
                name = i['name']
                h_data['Name'].append(name)
                description = i['description']
                h_data['Description'].append(description)
                neighborhood = i['neighborhood_overview']
                h_data['Neighborhood_Overview'].append(neighborhood)
                trans = i['transit']
                h_data['Transit'].append(trans)
                stre = i['address']['street']
                h_data['Street'].append(stre)
                sub = i['address']['suburb']
                h_data['Suburb'].append(sub)
                gov_area =i['address']['government_area']
                h_data['Government_area'].append(gov_area)
                longitude = i['address']['location']['coordinates'][0]
                h_data['Longitude'].append(longitude)
                latitude= i['address']['location']['coordinates'][1]
                h_data['Latitude'].append(latitude)
                market = i['address']['market']
                h_data['Market'].append(market)
                country = i['address']['country']
                h_data['Country'].append(country)
                country_code =i['address']['country_code']
                h_data['Country_code'].append(country_code)
                price = i['price']
                h_data['Price'].append(price)
                securitydep = i.get('security_deposit')
                h_data['Security_deposit'].append(securitydep)
                weeklypr = i.get('weekly_price')
                h_data['Weekly_price'].append(weeklypr)
                monthlypr = i.get('monthly_price')
                h_data['Monthly_price'].append(monthlypr)
                cleaningfee = i.get('cleaning_fee')
                h_data['Cleaning_fee'].append(cleaningfee) 
                cancellp = i['cancellation_policy']
                h_data['Cancellation_policy'].append( cancellp)
        

        hotel_data=pd.DataFrame(h_data)            #dataframe

        hotel_data['Security_deposit']=hotel_data['Security_deposit'].fillna(0)
        hotel_data['Weekly_price']=hotel_data['Weekly_price'].fillna(0)
        hotel_data['Monthly_price']=hotel_data['Monthly_price'].fillna(0)
        hotel_data['Cleaning_fee']=hotel_data['Cleaning_fee'].fillna(0)
        hotel_data['Price']=hotel_data['Price'].astype(dtype='int64')
        hotel_data['Security_deposit']=hotel_data['Security_deposit'].astype(dtype='int64')
        hotel_data['Weekly_price']=hotel_data['Weekly_price'].astype(dtype='int64')
        hotel_data['Monthly_price']=hotel_data['Monthly_price'].astype(dtype='int64')
        hotel_data['Cleaning_fee']=hotel_data['Cleaning_fee'].astype(dtype='int64')


        return hotel_data

hotel_data=hotel_info(airb_data)
#hotel_data=pd.DataFrame(h_data)     

```

```python
#Host details:
def host_info():
    ho_data ={'id':[],'Host_id':[],'Host_name':[],'Host_location':[],'Host_about':[],'Host_verifications':[],
    'Host_response_rate':[],'Host_neighbourhood':[],'Host_total_listings':[]}

    for i in airb_data:
        id = i['_id']
        ho_data['id'].append(id)
        hostid = i['host']['host_id']
        ho_data['Host_id'].append(hostid)
        hostname = i['host']['host_name']
        ho_data['Host_name'].append(hostname)
        hostloc = i['host']['host_location']
        ho_data['Host_location'].append(hostloc)
        hostabt = i['host']['host_about']
        ho_data['Host_about'].append(hostabt)
        hostver = i['host']['host_verifications']
        ho_data['Host_verifications'].append(hostver)
        hostrate = i['host'].get('host_response_rate')
        ho_data['Host_response_rate'].append(hostrate)
        hostnei = i['host']['host_neighbourhood']
        ho_data['Host_neighbourhood'].append(hostnei)
        total_lt = i['host']['host_total_listings_count']
        ho_data['Host_total_listings'].append(total_lt)

    host_data=pd.DataFrame(ho_data)               #dataframe

    host_data['Host_response_rate']=host_data['Host_response_rate'].fillna(0)
    host_data['Host_response_rate']=host_data['Host_response_rate'].astype(dtype='int64')
    host_data['Host_verifications']=[','.join(map(str,x)) if isinstance else ' '  for x in host_data['Host_verifications']]

    return host_data

host_data=host_info()
#host_data=pd.DataFrame(ho_data)
```

```python
#room details:
def room_info():

    ro_data = {'id':[],'Images':[],'Property_type':[],'Room_type':[],'Bed_Type':[],'Accomodates':[],
    'Maximum_nights':[],'Minimum_nights':[],'Bedrooms':[],'Beds':[],'Bathrooms':[],'Guest_included':[],
    'Amenities':[],'Availability_30':[],'Availability_60':[],'Availability_90':[],'Availability_365':[]}

    for i in airb_data:
        id = i['_id']
        ro_data['id'].append(id)
        image = i['images']['picture_url']
        ro_data['Images'].append(image)
        property = i['property_type']
        ro_data['Property_type'].append(property)
        room = i['room_type']
        ro_data['Room_type'].append(room)
        bed= i['bed_type']
        ro_data['Bed_Type'].append(bed)
        accom = i['accommodates']
        ro_data['Accomodates'].append(accom)
        max = i['maximum_nights']
        ro_data['Maximum_nights'].append(max)
        min = i['minimum_nights']
        ro_data['Minimum_nights'].append(min)
        bedrm = i.get('bedrooms')
        ro_data['Bedrooms'].append(bedrm)
        beds = i.get('beds')
        ro_data['Beds'].append(beds)
        bathrm = i.get('bathrooms')
        ro_data['Bathrooms'].append(bathrm )
        guestin = i.get('guests_included')
        ro_data['Guest_included'].append(guestin)
        amment = i['amenities']
        ro_data['Amenities'].append(amment)
        avail30 = i['availability']['availability_30']
        ro_data['Availability_30'].append(avail30)
        avail60 = i['availability']['availability_60']
        ro_data['Availability_60'].append(avail60)
        avail90 = i['availability']['availability_90']
        ro_data['Availability_90'].append(avail90)
        avail365 = i['availability']['availability_365']
        ro_data['Availability_365'].append(avail365)

    room_data=pd.DataFrame(ro_data)                    #dataframe
    
    room_data['Beds']=room_data['Beds'].fillna(0)
    room_data['Bedrooms']=room_data['Bedrooms'].fillna(0)
    room_data['Bathrooms']=room_data['Bathrooms'].fillna(0)
    room_data['Beds']=room_data['Beds'].astype(dtype='int64')
    room_data['Bedrooms']=room_data['Bedrooms'].astype(dtype='int64')
    room_data['Bathrooms']=room_data['Bathrooms'].astype(dtype='int64')
    room_data['Guest_included']=room_data['Guest_included'].astype(dtype='int64')
    room_data['Amenities']=[', '.join(str(a) for a in x) if isinstance(x, list) else ' ' for x in room_data['Amenities']]

    return room_data

room_data=room_info()
#room_data=pd.DataFrame(ro_data)     

```

```python
#review details:
def review_info():
    re_data={'id':[],'Number_of_review':[],'First_review':[],'Last_review':[],'Overall_score':[],'Cleanliness_score':[],
    'Checking_score':[],'Communication_score':[],'Rating':[],'Location_score':[]}

    for i in airb_data:
        id = i['_id']
        re_data['id'].append(id)
        review = i['number_of_reviews']
        re_data['Number_of_review'].append(review)
        firstr = i.get('first_review')
        re_data['First_review'].append(firstr)
        lastr = i.get('last_review')
        re_data['Last_review'].append(lastr)
        overscore = i['review_scores'].get('review_score_accuracy')
        re_data['Overall_score'].append(overscore)
        cleanscore = i['review_scores'].get('review_score_cleanliness')
        re_data['Cleanliness_score'].append(cleanscore)
        checckscore = i['review_scores'].get('review_score_checkin')
        re_data['Checking_score'].append(checckscore)
        commscore = i['review_scores'].get('review_score_communication')
        re_data['Communication_score'].append(commscore)
        rating = i['review_scores'].get('review_score_rating')
        re_data['Rating'].append(rating)
        loc = i['review_scores'].get('review_score_location')
        re_data['Location_score'].append(loc)

    review_data=pd.DataFrame(re_data)       #dataframe

    review_data['First_review']=review_data['First_review'].fillna('Not avaiable')
    review_data['Last_review']=review_data['Last_review'].fillna('Not avaiable')
    review_data['Cleanliness_score']=review_data['Cleanliness_score'].fillna(0)
    review_data['Overall_score']=review_data['Overall_score'].fillna(0)
    review_data['Checking_score']=review_data['Checking_score'].fillna(0)
    review_data['Communication_score']=review_data['Communication_score'].fillna(0)
    review_data['Rating']=review_data['Rating'].fillna(0)
    review_data['Location_score']=review_data['Location_score'].fillna(0)
    
    return review_data


#review_data=pd.DataFrame(re_data)
review_data=review_info()
```

```python
#reviewer details:
def reviewer_info():
    rv_data = {'id':[],'Comment_id':[],'Date':[],'Reviewer_id':[],'Reviewer_name':[],'Comments':[]}

    for r in airb_data:
        j_list = r.get('reviews',[])
        for i in j_list:
            id = i.get('listing_id')
            rv_data['id'].append(id)
            commentid = i.get('_id')
            rv_data['Comment_id'].append(commentid)
            date = i.get('date')
            rv_data['Date'].append(date)
            reviewid = i.get('reviewer_id')
            rv_data['Reviewer_id'].append(reviewid)
            reviewname = i.get('reviewer_name')
            rv_data['Reviewer_name'].append(reviewname)
            comment = i.get('comments')
            rv_data['Comments'].append(comment)

    reviewer_data=pd.DataFrame(rv_data)           #dataframe
    
    reviewer_data['Reviewer_name']=reviewer_data['Reviewer_name'].fillna('Not Available')
    reviewer_data['Comments']=reviewer_data['Comments'].fillna('Not Available')
    
    return reviewer_data

reviewer_data=reviewer_info()
#reviewer_data=pd.DataFrame(rv_data)
```

```python
#Correlation Heatmap Visualization
def corr_info():
    c_data = {'Listing_id':[], 'Host_id':[],'Longitude':[],'Latitude':[],'Price':[],'Number_of_review':[],
            'Minimum_nights':[],'Availability_365':[],'Host_response_rate':[]}

    for i in airb_data:
        ids = i['_id']
        c_data['Listing_id'].append(ids)
        hostid = i['host']['host_id']
        c_data['Host_id'].append(hostid)
        longitude = i['address']['location']['coordinates'][0]
        c_data['Longitude'].append(longitude)
        latitude= i['address']['location']['coordinates'][1]
        c_data['Latitude'].append(latitude)
        price = i['price']
        c_data['Price'].append(price)
        review = i['number_of_reviews']
        c_data['Number_of_review'].append(review)
        min = i['minimum_nights']
        c_data['Minimum_nights'].append(min)
        avail365 = i['availability']['availability_365']
        c_data['Availability_365'].append(avail365)
        hostrate = i['host'].get('host_response_rate')
        c_data['Host_response_rate'].append(hostrate)
        
    corr_data=pd.DataFrame(c_data) 

    corr_data['Price']=corr_data['Price'].astype(dtype='int64')
    corr_data['Host_response_rate']=corr_data['Host_response_rate'].fillna(0)
    corr_data['Host_response_rate']=corr_data['Host_response_rate'].astype(dtype='int64')

    return corr_data

corr_data=corr_info()

```

```python
#minimum_night:
def min_n():
    mycursor.execute('''SELECT hd.Price, SUM(Minimum_nights) as Min_night FROM airb.room_details ro JOIN airb.hotel_details hd ON ro.id = hd.id GROUP BY Price;''')
    a15=mycursor.fetchall()
    b15=pd.DataFrame(a15,columns=("Price","Min_night"))
    return b15

min_st=min_n()
```

```python
# price vs the minimum_nights:
plt.figure(figsize=(10, 8))
sns.scatterplot(min_st, x='Price', y='Min_night', hue='Price', palette='coolwarm', alpha=0.7)
plt.xlabel('Price')
plt.ylabel('Min_night')
plt.title('Price vs. Minimum_nights')
plt.show()

#fig.shown-->fig3.png
```
